```python
!pip install numpy matplotlib



import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros iniciales
G = 6.67430e-11  # Constante gravitacional
M = 5.972e24     # Masa de la Tierra
dt = 60          # Paso de tiempo en segundos
num_steps = 1000 # Número de pasos de tiempo

# Vectores de posición y velocidad iniciales
initial_position = np.array([1.496e11, 0, 0], dtype=float)   # Vector de posición inicial
initial_velocity = np.array([0, 29800, 0], dtype=float)      # Vector de velocidad inicial

# Función de propagación de órbita de Cowell
def propagate_orbit(position, velocity, dt, num_steps):
    positions = [position]  # Lista para almacenar los vectores de posición en cada paso
    for _ in range(num_steps):
        # Calcula la aceleración gravitacional
        acceleration = -G * M * position / np.linalg.norm(position)**3
        # Actualiza la velocidad y la posición utilizando el método de Euler
        velocity += acceleration * dt
        position += velocity * dt
        # Agrega el nuevo vector de posición a la lista
        positions.append(position)
    return np.array(positions)

# Propaga la órbita
orbit = propagate_orbit(initial_position, initial_velocity, dt, num_steps)

# Grafica la órbita en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(orbit[:, 0], orbit[:, 1], orbit[:, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

```
