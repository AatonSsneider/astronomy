```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
G = 6.67430e-11  # Constante de gravitación universal
M_sol = 1.989e30  # Masa del Sol
M_tierra = 5.972e24  # Masa de la Tierra
R_tierra = 1.496e11  # Distancia promedio de la Tierra al Sol
V_tierra = 2.9783e4  # Velocidad orbital de la Tierra

# Función para calcular la aceleración gravitacional
def aceleracion_gravitacional(r):
    magnitud = G * M_sol / np.linalg.norm(r)**3
    return -magnitud * r

# Simulación de la órbita de la Tierra
dt = 60  # Paso de tiempo en segundos
num_pasos = 365 * 24 * 60 * 60 // dt  # Número de pasos en un año
posicion = np.zeros((num_pasos, 2))
velocidad = np.zeros((num_pasos, 2))

posicion[0] = [R_tierra, 0]  # Posición inicial
velocidad[0] = [0, V_tierra]  # Velocidad inicial

for i in range(1, num_pasos):
    # Actualizar la posición y velocidad utilizando el método de Euler
    aceleracion = aceleracion_gravitacional(posicion[i-1])
    posicion[i] = posicion[i-1] + velocidad[i-1] * dt
    velocidad[i] = velocidad[i-1] + aceleracion * dt

# Graficar la órbita de la Tierra
plt.plot(posicion[:, 0], posicion[:, 1])
plt.axis('equal')
plt.xlabel('Distancia (m)')
plt.ylabel('Distancia (m)')
plt.title('Órbita de la Tierra')
plt.grid(True)
plt.show()
```
